# Stock Market Prediction using LSTM

This notebook demonstrates how to predict stock market prices using **Long Short-Term Memory (LSTM)** networks. We will use stock data from the **Nairobi Stock Exchange (NSE)** (or any other available data). In this example, we will predict stock prices using historical **closing prices**.

## Steps:
1. **Load and preprocess the stock data**
2. **Build and train the LSTM model**
3. **Evaluate the model's performance**
4. **Visualize actual vs predicted prices**


### Step 1: Import Libraries and Load Stock Data
First, we will import necessary libraries and load the stock data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
# Download stock data from dataset. Note that this is data from 5 years ago 
# and does not reflect the current stock market
data = pd.read_csv('Datasets/NSE_SCOM_Safaricom.csv')

# Display first 5 rows of the data
data.head()
```

### Step 2: Data Preprocessing
We will focus on the 'Close' price because it is commonly used for stock price prediction. Additionally, we will **normalize** the prices using the `MinMaxScaler` to scale the data between 0 and 1, which is ideal for LSTM models.

```python
# Focus on the 'Close' price
data = data[['Close']]
```

```python
# Visualize the stock prices
plt.figure(figsize=(12, 6))
plt.plot(data['Close'], label='Close Price')
plt.title(f'Stock Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
# Normalize the stock prices using MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)
```

```python
# Prepare data for LSTM (using 60 timesteps to predict the next day's price)
X = []
y = []
for i in range(60, len(scaled_data)):
    X.append(scaled_data[i-60:i, 0])
    y.append(scaled_data[i, 0])
```

```python
# Convert X and y to numpy arrays
X, y = np.array(X), np.array(y)
```

```python
# Reshape X to be 3D for LSTM input: [samples, time steps, features]
X = np.reshape(X, (X.shape[0], X.shape[1], 1))
```

### Step 3: Build and Train the LSTM Model
Next, we will build the LSTM model with 2 LSTM layers and Dropout layers for regularization. Finally, we will compile the model using the Adam optimizer and Mean Squared Error loss function.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
```

```python
# Build the LSTM model
model = Sequential()
```

```python
# First LSTM Layer with Dropout for regularization
model.add(LSTM(units=50, return_sequences=True, input_shape=(X.shape[1], 1)))
model.add(Dropout(0.2))
```

```python
# Second LSTM Layer with Dropout
model.add(LSTM(units=50, return_sequences=False))
model.add(Dropout(0.2))
```

```python
# Fully connected output layer
model.add(Dense(units=1))
```

```python
# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Train the model
model.fit(X, y, epochs=10, batch_size=32)
```

### Step 4: Make Predictions and Visualize Results
After training the model, we will make predictions on the test set (the last 100 days of the stock data). We will then visualize the actual vs predicted stock prices.

```python
# Predict stock prices on the test set
test_data = data.tail(100)  # Using last 100 days for prediction
scaled_test_data = scaler.transform(test_data)
```

```python
X_test = []
for i in range(60, len(scaled_test_data)):
    X_test.append(scaled_test_data[i-60:i, 0])
```

```python
X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
```

```python
# Make predictions
predicted_prices = model.predict(X_test)
```

```python
# Invert scaling for actual prices
predicted_prices = scaler.inverse_transform(predicted_prices)
```

```python
# Visualize the results
plt.figure(figsize=(12, 6))
plt.plot(data.index[-100:], test_data['Close'], color='blue', label='Actual Stock Price')
plt.plot(data.index[-40:], predicted_prices, color='red', label='Predicted Stock Price')
plt.title(f'Stock Price Prediction')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

### Step 5: Model Evaluation
We will now calculate the **Root Mean Squared Error (RMSE)** to assess the accuracy of our predictions. A lower RMSE indicates better performance.

```python
from sklearn.metrics import mean_squared_error

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(test_data['Close'][60:], predicted_prices))
print(f'Root Mean Squared Error (RMSE): {rmse:.4f}')
```
